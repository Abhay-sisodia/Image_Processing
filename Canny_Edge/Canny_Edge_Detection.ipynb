```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('nature.jpg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#  Now we will try and implement canny edge detection 
img=cv2.medianBlur(img,5)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sobel_x=cv2.Sobel(img,-1,1,0)
sobel_y=cv2.Sobel(img,-1,0,1)
sobel_xy=cv2.bitwise_or(sobel_x,sobel_y)
sobel_final=cv2.addWeighted(sobel_x,0.5,sobel_y,0.5,0)
laplacian=cv2.Laplacian(img,-1)


cv2.namedWindow('Image_show',cv2.WINDOW_NORMAL)
cv2.resizeWindow('Image_show',(600,600))
cv2.imshow('Image_show',sobel_x)
cv2.waitKey(0)
cv2.imshow('Image_show',sobel_y)
cv2.waitKey(0)
cv2.imshow('Image_show',sobel_xy)
cv2.waitKey(0)
cv2.imshow('Image_show',sobel_final)
cv2.waitKey(0)
cv2.imshow('Image_show',laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('Canny',cv2.WINDOW_NORMAL)
cv2.resizeWindow('Canny',(600,600))
canny_output=cv2.Canny(img,100,240)
cv2.imshow('Canny',canny_output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
