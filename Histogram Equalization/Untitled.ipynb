```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('nature.jpg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
# Plotting histogram for each channel of the image
b,g,r=cv2.split(img)
hist=cv2.calcHist([b],[0],None,[256],[0,255])
plt.plot(hist,color='b')
hist=cv2.calcHist([g],[0],None,[256],[0,255])
plt.plot(hist,color='g')
hist=cv2.calcHist([r],[0],None,[256],[0,255])
plt.plot(hist,color='r')
plt.title('Histogram for original image')
```

```python
# Applying histogram equalization to image
img1=img.copy()
b,g,r=cv2.split(img1)
b=cv2.equalizeHist(b)
g=cv2.equalizeHist(g)
r=cv2.equalizeHist(r)
img_eq=cv2.merge((r,g,b))
plt.imshow(img_eq)
```

```python
# Histogram equalization Plot for the equalised bgr image formed
hist=cv2.calcHist([b],[0],None,[256],[0,255])
plt.plot(hist,color='b')
hist=cv2.calcHist([g],[0],None,[256],[0,255])
plt.plot(hist,color='g')
hist=cv2.calcHist([r],[0],None,[256],[0,255])
plt.plot(hist,color='r')
plt.title('Histogram Equalization')
```

```python
# Applying Histogram Equalization wrt to hsv values
img2=img.copy()
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(img2)
# Here we only tend to equalise values
v=cv2.equalizeHist(v)
img2_new=cv2.merge((h,s,v))
img2_new=cv2.cvtColor(img2_new,cv2.COLOR_HSV2RGB)
plt.imshow(img2_new)
plt.title('Histogram Equalization wrt hsv')
```

```python
hist=cv2.calcHist([h],[0],None,[181],[0,180])
plt.plot(hist,color='b')
hist=cv2.calcHist([s],[0],None,[256],[0,255])
plt.plot(hist,color='g')
hist=cv2.calcHist([v],[0],None,[256],[0,255])
plt.plot(hist,color='red')
plt.title('Histogram Equalization in hsv')
```

```python
# Applying Adaptive histogram equalization in image
#  Clahe works only for grayscale images
img3=img.copy()
img3=cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)
clahe=cv2.createCLAHE(clipLimit=5)
img3_new=clahe.apply(img3)
plt.imshow(img3_new,cmap='gray')
```

```python
fig,ax=plt.subplots(2,2,figsize=(16,8))
ax[0][0].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax[0][0].set_title('Original Image')
ax[0][1].imshow(img_eq)
ax[0][1].set_title('HE Image (RGB)')
ax[1][0].imshow(img2_new)
ax[1][0].set_title('HE Image (HSV)')
ax[1][1].imshow(img3_new,cmap='gray')
ax[1][1].set_title('CLAHE Image')
```
